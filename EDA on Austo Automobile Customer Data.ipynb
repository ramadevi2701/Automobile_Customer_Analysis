# **BIVARIATE ANALYSIS**

**Explore the relationship between all numerical variables**

```python
numerical_columns = Customer_Data.select_dtypes(include=['number']).columns
plt.figure(figsize=(10,10))
for i in range(len(numerical_columns)-1):
    col1 = numerical_columns[i]
    col2 = numerical_columns[i+1]
    plt.subplot(3, 2, i+1)
    sns.scatterplot(x=Customer_Data[col1], y=Customer_Data[col2])
    plt.title(f"Scatter Plot of {col1} vs {col2}")
    plt.xlabel(col1)
    plt.ylabel(col2)
plt.tight_layout()
plt.show()
```

Although the initial bivariate analysis was performed on all numerical variables, further insights relevant to the marketing campaign required focusing on specific variable combinations. Therefore, we selected only the columns that directly influence customer purchasing behavior, spending capacity, and car preference. This targeted approach ensures that the analysis remains aligned with the business objective and provides actionable insights for marketing decisions.

```python
# Define important numerical pairs for bivariate analysis
important_pairs = [
    ('Salary', 'Price'),
    ('Total_salary', 'Price'),
    ('Age', 'Price'),
    ('Salary', 'Partner_salary')
]
plt.figure(figsize=(12, 10))
# Colors for each scatterplot
colors = ['teal', 'purple', 'green', 'orange']
# Loop through pairs
for i, (xcol, ycol) in enumerate(important_pairs, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=Customer_Data[xcol],
                    y=Customer_Data[ycol],
                    color=colors[i-1],
                    alpha=0.7)
    plt.title(f"{xcol} vs {ycol}")
    plt.xlabel(xcol)
    plt.ylabel(ycol)
plt.tight_layout()
plt.show()
```

**Observation and Insights**

* Salary vs Price

The scatterplot shows a weak positive trend. Customers with higher salaries tend to buy slightly more expensive cars, but the points are widely spread. This means salary influences car price, but it is not the only factor. Other elements like personal preferences, loans, or family needs may also affect the purchase decision.

* Total_salary vs Price

This plot shows a clearer positive relationship compared to individual salary. As total household income increases, the price of the vehicle purchased also tends to increase. Higher-income families have a broader price range, meaning they can afford both mid-range and premium cars. This makes total household income a stronger predictor of purchasing power.

* Age vs Price

There is no strong visible pattern in this scatterplot. Customers of different ages (25–55) seem to purchase cars across similar price ranges. This indicates that age does not significantly influence the price of the car selected. Both younger and older individuals buy mid-range as well as higher-priced cars.

* Salary vs Partner_salary

The scatterplot shows mostly horizontal clustering, meaning partner salary remains low for many customers regardless of customer salary. This suggests that many households are single-income, or partner income does not scale with the customer's income. There is no strong relationship between the two variables.

# **ACTIONABLE INSIGHTS AND RECOMMENDATIONS**

**Actionable Insights**

1. Car preferences vary strongly by gender

From the Make vs Gender analysis:

*  Females prefer SUVs more than males, as shown by significantly higher female SUV counts.

*  Males overwhelmingly prefer Sedans, with the highest male count in the Sedan category.

*  Hatchbacks are purchased mostly by males, with minimal female involvement.

Implication: Gender plays an important role in car type preference.

2. Profession significantly influences car choice
From the Make vs Profession plot:

*  Salaried individuals strongly prefer Sedans, followed by Hatchbacks.

*  Business professionals prefer Sedans and Hatchbacks almost equally, with very low SUV preference.

Implication: Sedans dominate across professions, but SUVs are less attractive to business professionals compared to salaried females.

3. Salaried males are NOT easier targets for SUV sales
From the multivariate breakdown:

*  Salaried males show the highest count for Sedans, not SUVs.

*  Female salaried customers show more interest in SUVs.

Therefore, the assumption that a “salaried male is an easier target for SUV sales” is not supported by the data.

Implication: Marketing SUV campaigns toward salaried males would be ineffective.

4. Female customers spend slightly more on average
From the Average Price vs Gender plot:

*  Females spend an average of ₹47,705, which is higher than the male average.

This suggests that female buyers may opt for slightly higher-priced or feature-rich models.

Implication: Female customers represent a high-value segment for premium offerings.

5. Individuals who took personal loans spent slightly less
From the Personal Loan vs Price (sum) plot:

*  Customers without loans spent more in total than those who took loans.

This indicates loan-takers may be more price-sensitive or opt for budget-friendly models.

Implication: Loan-takers require more flexible finance schemes to encourage higher-value purchases.

6. Working partner status has little to no impact on spending
From the Working Partner vs Price plot:

*  Average car price is almost identical for customers with or without a working partner.

This indicates dual-income households do not significantly change car spending behavior in this dataset.

Implication: Partner's working status is not a strong predictor of high-value purchases.

7. Hatchback purchases are driven mainly by males
Across all segments:

*  Hatchback is dominated by male customers.

*  Female hatchback purchases are extremely low.

Implication: Hatchbacks appeal to practical, possibly first-time male buyers.

**Business Recommendations**

1. Position SUVs in women-focused marketing campaigns,
Since female customers show stronger SUV preference:

*  Create women-centric promotional offers.

*  Highlight features like safety, comfort, and family suitability.

*  Use influencer marketing targeting female audiences.

Outcome: Higher SUV conversions among female buyers.

2. Focus Sedan promotions on salaried and male customers,
Since Sedans are most preferred by males and salaried professionals:

*  Provide special corporate EMI plans for salaried employees.

*  Highlight performance, reliability, and comfort features.

*  Offer loyalty discounts for professionally employed customers.

Outcome: Boost sedan sales in the strongest demand segment.

3. Promote premium upgrades to female buyers,
Because females spend more on average:

*  Promote automatic variants, safety packages, and premium trims.

*  Provide trial drives with premium models.

*  Offer women-exclusive financing options.

Outcome: Higher revenue from premium product lines.

4. Create targeted finance schemes for loan-takers,
Loan-takers tend to spend less:

*  Introduce lower-interest auto loan packages.

*  Provide bundled insurance + accessory discounts.

*  Encourage “Upgrade Now, Pay Later” schemes.

Outcome: Increased spending and improved customer conversion.

5. Avoid using working-partner status as a segmentation strategy,
The dataset shows no meaningful difference in spending between:

*  Customers with a working partner

*  Customers without one

Therefore:

*  Do not allocate marketing budget to “dual-income” targeting.

*  Focus instead on gender, profession, and loan status which are far stronger predictors.

6. Strengthen marketing of Hatchbacks toward young male professionals,
Since hatchbacks show male dominance:

*  Market a value-for-money positioning.

*  Highlight mileage, maintenance cost, and compact design.

*  Offer budget-friendly EMI plans for young customers.

Outcome: Improved penetration in the male entry-level segment.

7. Use profession-based segmentation in advertising,
Profession clearly influences car type choice:

*  Promote Sedans in corporate newsletters, LinkedIn ads, and IT park events.

*  Promote SUVs and Hatchbacks through entrepreneurship communities and small business networks.

Outcome: More relevant targeting and higher engagement.
